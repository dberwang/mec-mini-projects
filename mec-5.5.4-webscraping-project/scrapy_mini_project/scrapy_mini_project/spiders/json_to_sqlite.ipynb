```python
import json
import pandas as pd

# Open and load our css json file into a dataframe
with open('css-scraper-results.json') as f:
    data_css = json.load(f)
df_css = pd.DataFrame(data_css)

# Clean and view
df_css.rename(columns={"text": "Quote", "author": "Author"}, inplace=True)
df_css['Quote'] = df_css['Quote'].str.replace('[^\x00-\x7F]','')
df_css['tags'] = [','.join(map(str, l)) for l in df_css['tags']]
df_css
```

```python
# Use sqlalchemy to create a sqlite database from the dataframe
from sqlalchemy import create_engine

engine = create_engine('sqlite:///css.db', echo=True)
sqlite_connection = engine.connect()
sqlite_table = "quotes"
df_css.to_sql(sqlite_table, sqlite_connection, if_exists='fail')
```
